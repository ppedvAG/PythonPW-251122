### N.A.

Keine automatische LÃ¶sung -> manuell

Hier wird die string Replace Funktion verwendet

```python
data == "N.A."
```

```python
(data == "N.A.").value_counts()
```

Die Fehler mÃ¼ssen in 3 Spalten behoben werden -> Funktion

```python
def fixNA(columnName):
    booleanMaske = data[columnName] == "N.A."
    fehlerZeilen = data[booleanMaske]
    fehlerZeilen[columnName] = 0
    data.loc[fehlerZeilen.index] = fehlerZeilen
```

```python
data
```

```python
fixNA("MedAge")  # Einzelne Spalte reparieren
```

```python
for col in data.columns:  # Alle Spalten reparieren
    fixNA(col)
```

```python
data
```

### Prozentzeichen

Hier string Replace

```python
data["YearlyChange"].replace(" %", "", regex=True)  # Suche nach " %" und ersetze es durch ""
```

```python
repariert = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data["YearlyChange"] = repariert
```

```python
data
```

```python
def fixPct(columnName):
    r = data[columnName].replace(" %", "", regex=True)
    data[columnName] = r
```

```python
fixPct("UrbanPop")  # Einzelne Spalte reparieren
```

```python
for col in data.columns:  # Alle Spalten reparieren
    fixPct(col)
```

```python
data
```
