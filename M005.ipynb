### Globaler vs. Objektorientierter Modus

Globaler Modus: Befehle ausfÃ¼hren Ã¼ber plt, mit plt.show() die Konfiguration abschlieÃŸen

Objektorientierter Modus: Figure Objekte und Axis Objekte werden dediziert angesprochen, und dadurch konfiguriert

```python
plt.plot(pop["Population"])
plt.grid(alpha=0.2)
plt.show()  # SchlieÃŸt die Konfiguration des Plots ab, und zeigt ihn an

plt.plot(pop["WorldSharePct"])
plt.show()
```

```python
plt.figure(figsize=(12, 3))
plt.plot(pop["Population"])
plt.show()
```

```python
fig, ((ax1_1, ax1_2), (ax2_1, ax2_2)) = plt.subplots(2, 2)
ax1_1.plot(np.random.random(100))
ax1_2.pie(pop["WorldSharePct"])
ax2_1.bar(pop.index, pop["Population"])
ax2_2.hist(np.random.randint(100, size=100), bins=20)
fig.savefig("Test.png")
None
```
